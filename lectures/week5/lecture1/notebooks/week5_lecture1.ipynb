# 7. Lynx Trappings

```python
lynx = pd.read_csv('Annual_Canadian_Lynx_trappings.csv', index_col=0)
lynx.head()
```

```python
# Write your code here.
plt.figure(figsize=(10, 5))
plt.title('Annual Lynx Trappings in Canada', fontsize=18)
ax = sns.lineplot(data=lynx, x='time', y='value', marker='o')
ax.xaxis.set_tick_params(labelsize=14)
ax.yaxis.set_tick_params(labelsize=14)
ax.set_xlabel('Date-Time', fontsize=18)
ax.set_ylabel('Annual Trappings', fontsize=18)
plt.show()
```

```python
lag = 11

fig, axs = plt.subplots(nrows=3, figsize=(8, 12))
axs[0].set_title('Autocorrelation Lag: {}'.format(lag), fontsize=16)
axs[0].plot(lynx['time'], lynx['value'], '-o', ms=10);
axs[0].plot(lynx['time']+lag, lynx['value'], '-o');
axs[0].set_xlim([1821, 1934])
axs[0].set_ylabel('Annual Trappings', fontsize=14)
axs[0].set_xlabel('Date-Time', fontsize=14)

corr = np.corrcoef(lynx['value'].iloc[0:-lag], lynx['value'].iloc[lag:])[0, 1]
axs[1].set_title('Correlation: {}'.format(np.round(corr, 2)), fontsize=16)
axs[1].plot(lynx['value'].iloc[0:-lag], lynx['value'].iloc[lag:], '.');
axs[1].set_aspect('equal', 'box')
axs[1].set_xlabel('Annual Trappings', fontsize=14)
axs[1].set_ylabel('Annual Trappings (lag: {})'.format(lag), fontsize=14)

lags = np.arange(0, lag+1)
corrs = [1]
corrs.extend([np.corrcoef(lynx['value'].iloc[0:-lag], lynx['value'].iloc[lag:])[0, 1] for lag in lags[1:]])
axs[2].set_xlabel('Lag', fontsize=14)
axs[2].set_ylabel('Autocorrelation', fontsize=14)
axs[2].set_xlim([0, 11])
axs[2].set_ylim([-1, 1])

axs[2].plot(lags, corrs, '-o')

# Narrow the gap between the plots
plt.subplots_adjust(hspace=0.35)
```

```python
pd.plotting.autocorrelation_plot(lynx['value']);
```

# 10. Red Meat Slaughter

```python
red_meat = pd.read_csv('monthly_red_meat_slaughter.csv')
red_meat['EndDt_DtFin'] = pd.to_datetime(red_meat['EndDt_DtFin'])
red_meat.head()
```

```python
cattle = red_meat[red_meat['MjrCmdtyEn_PrdtPrncplAn'] == 'Cattle']
cattle = cattle.set_index('EndDt_DtFin')
cattle = cattle.groupby(cattle.index).sum()
cattle.head()
```

```python
# Write your code here.
plt.figure(figsize=(10, 5))
plt.title('Monthly Cattle Slaughtered in Canada', fontsize=18)
ax = sns.lineplot(data=cattle, x=cattle.index, y='NumHd_NmbTetes', marker='o')
ax.xaxis.set_tick_params(labelsize=14)
ax.yaxis.set_tick_params(labelsize=14)
ax.set_xlim([datetime.date(1998, 1, 1), datetime.date(2003, 1, 1)])
ax.set_ylim([240000, 380000])
ax.set_xlabel('Date-Time', fontsize=18)
ax.set_ylabel('Cattle Slaughtered', fontsize=18)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
pd.plotting.autocorrelation_plot(cattle.loc['1998-01-01':'2003-01-01']);
```
